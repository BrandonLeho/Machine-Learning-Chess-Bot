```python
import chess.pgn

input_pgn_file = '/content/drive/My Drive/lichess_db_100k.pgn'
output_pgn_file = '/content/drive/My Drive/lichess_db_100k_filtered.pgn'

min_elo = 1500

games_to_keep = []
with open(input_pgn_file) as pgn_file:
    while True:
        game = chess.pgn.read_game(pgn_file)
        if game is None:
            break

        white_elo = game.headers.get('WhiteElo', '0')
        black_elo = game.headers.get('BlackElo', '0')
        try:
            white_elo = int(white_elo)
            black_elo = int(black_elo)

            if white_elo >= min_elo and black_elo >= min_elo:
                games_to_keep.append(game)
        except ValueError:

            pass

with open(output_pgn_file, 'w') as output_file:
    for game in games_to_keep:
        output_file.write(str(game) + '\n\n')

```
